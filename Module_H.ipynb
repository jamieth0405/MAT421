22.1 ODE Initial Value Problem Statement

- An Ordinary Differential Equation (ODE) is a relationship involving a function, its derivatives, and one independent variable.

- An ODE of order $n$ can be written as:
$$F\left(x, f(x), \frac{df(x)}{dx}, \frac{d^2f(x)}{dx^2}, \dots, \frac{d^{n-1} f(x)}{dx^{n-1}} \right) = \frac{d^n f(x)}{dx^n}$$

- To solve an ODE of order $n$, we require initial values: $$f(0), f'(0), f''(0), \dots, f^{(n-1)}(0)$$

- A solution satisfying the differential equation and these values is called the **particular solution**, and the task of finding it is called solving an **Initial Value Problem**.

- Initial value problems are widely used in physics and engineering because, with appropriate conditions, they yield **unique** solutions.

22.2 Reduction of Order

- Higher-order ODEs can be rewritten as systems of **first-order** ODEs by defining a **state vector**.

- For an ODE of order $n$, we define: $$
S(t) = \begin{bmatrix}
f(t) \\
f'(t) \\
f''(t) \\
\vdots \\
f^{(n-1)}(t)
\end{bmatrix}
$$

- Then: $$
\frac{dS(t)}{dt} = \begin{bmatrix}
f'(t) \\
f''(t) \\
f^{(3)}(t) \\
\vdots \\
f^{(n)}(t)
\end{bmatrix}
\Rightarrow \frac{dS(t)}{dt} = F(t, S(t))
$$
where $F$ is a function that appropriately assembles the vector describing the derivative of the state.

- This transforms a single $n$-th order equation into a system of $n$ first-order equations.

22.3 The Euler Method

- The **Euler Method** is a numerical technique to solve first-order ordinary differential equations (ODEs) of the form:
$$\frac{dS(t)}{dt} = F(t, S(t))$$

- Given an initial value $S(t_0) = S_0$, we compute the approximate solution using: $$S(t_{j+1}) = S(t_j) + h \cdot F(t_j, S(t_j))$$ Where:
 - $h$ is the step size
 - $t_j = t_0 + jh$


- Example: Solve the ODE: $$\frac{df(t)}{dt} = e^{-t}, \quad f(0) = -1$$ Using the Explicit Euler Method with step size $h = 0.1$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = lambda t, s: np.exp(-t)
h = 0.1
t = np.arange(0, 1 + h, h)
s0 = -1

# Euler method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Plot results
plt.figure(figsize=(15, 8))
plt.plot(t, s, 'bo--', label='Euler Approximation')
plt.plot(t, -np.exp(-t), 'g', label='Exact Solution')
plt.title('Euler Method vs. Exact Solution for Simple ODE with h = 0.1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid(True)
plt.legend()
plt.show()
```
