20.1 Numerical Differentiation Problem Statement

- Numerical grid: evenly spaced points over the domain of a function.
- Step size: distance between adjacent points.
- In Python, numerical grids can be generated using the *linspace* function of numpy library.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range and number of points
x = np.linspace(0, 10, 11)  # Creates 11 points from 0 to 10 (numerical grid)
y = np.sin(x)  # Function values at the grid points

# Plot the numerical grid
plt.scatter(x, y, color = 'red', label = 'Grid Points')
plt.plot(x, y, linestyle = 'dashed', color = 'blue', label = 'sin(x)')

plt.xlabel('x')
plt.ylabel('f(x) = sin(x)')
plt.title('Sin(x) with Numerical Grid')
plt.legend()
plt.show()

# Print the grid points
print("Numerical Grid (x values):", x)
print("Function values (sin(x)):", y)
```

20.2 Finite Difference Approximating Derivatives

- **Derivatives Approximation**: The derivative of a function at a point is defined as the slope at that point, and finite difference approximations estimate this slope using values from neighboring points.
- **Forward Difference**: Uses the function values at the current and next point.
- **Backward Difference**: Uses the function values at the current and previous point.
- **Central Difference**: Uses function values at both the next and previous points, offering higher accuracy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return np.sin(x)  # Function: f(x) = sin(x)

def df_exact(x):
    return np.cos(x)  # Exact derivative: f'(x) = cos(x)

# Define grid and step size
h = 0.1  # Step size
x = np.arange(0, 2*np.pi, h)  # Numerical grid

# Compute function values
y = f(x)

# Forward Difference Approximation
forward_diff = (f(x + h) - f(x)) / h

# Backward Difference Approximation
backward_diff = (f(x) - f(x - h)) / h

# Central Difference Approximation
central_diff = (f(x + h) - f(x - h)) / (2 * h)

# Exact derivative values
exact_solution = df_exact(x)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(x, exact_solution, label = 'Exact Derivative (cos(x))', linewidth = 2, color = 'black')
plt.plot(x, forward_diff, '--', label = 'Forward Difference', alpha = 0.7)
plt.plot(x, backward_diff, '--', label = 'Backward Difference', alpha = 0.7)
plt.plot(x, central_diff, '--', label = 'Central Difference', alpha = 0.7)

plt.xlabel('x')
plt.ylabel("Derivative Approximation")
plt.title("Finite Difference Approximations of Derivative")
plt.legend()
plt.grid(True)
plt.show()
```

The graph shows that the central difference method gives a higher-accurate approximation that it fits nicely to the curve.

20.3 Approximating of Higher Order Derivatives

- Higher-order derivatives can be approximated using Taylor series expansion.
- Just as finite difference methods approximate first-order derivatives, similar techniques can be used for second-order 𝒇"(𝑥), third-order 𝒇'''(𝑥), and beyond.
- Finite Difference Approximation for Second Derivative with Taylor series: $$
f''(x_j) \approx \frac{f(x_{j+1}) - 2f(x_j) + f(x_{j-1})}{h^2} $$
- This method has an accuracy of $O(h^2)$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its second derivative
def f(x):
    return np.sin(x)  # Function: f(x) = sin(x)

def d2f_exact(x):
    return -np.sin(x)  # Exact second derivative: f''(x) = -sin(x)

# Define grid and step size
h = 0.1  # Step size
x = np.arange(0, 2*np.pi, h)  # Numerical grid

# Compute function values
y = f(x)

# Second derivative using central finite difference formula:
# f''(x) ≈ (f(x+h) - 2f(x) + f(x-h)) / h^2
second_derivative = (f(x + h) - 2 * f(x) + f(x - h)) / h**2

# Exact second derivative
exact_solution = d2f_exact(x)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(x, exact_solution, label = "Exact Second Derivative (-sin(x))", linewidth = 2, color = 'pink')
plt.plot(x, second_derivative, '--', label = "Finite Difference Approximation", alpha = 0.7)

plt.xlabel("x")
plt.ylabel("Second Derivative Approximation")
plt.title("Finite Difference Approximation of Second Derivative")
plt.legend()
plt.grid(True)
plt.show()
```
